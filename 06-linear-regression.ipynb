```python
# Importando as bibliotecas necessárias
import numpy as np  # Para manipulação de arrays
from sklearn.linear_model import LinearRegression  # Implementação do modelo de regressão linear
import matplotlib.pyplot as plt  # Para visualização de dados
```

# Exemplo 1: Regressão linear simples - relação entre litros de combustível e quilômetros rodados

```python
# Criando os dados de entrada: x representa litros de combustível
# reshape((-1, 1)) transforma o array em uma matriz de uma coluna, necessário para o sklearn
x = np.array([2, 3, 5, 9, 12, 14, 18]).reshape((-1, 1))
# y representa quilômetros rodados
y = np.array([20, 40, 80, 120, 160, 200, 250])
# Criando o modelo de regressão linear e treinando com os dados
model = LinearRegression().fit(x, y)
```

```python
# Criando novos dados para predição (20, 25 e 30 litros de combustível)
new = np.array([20, 25, 30]).reshape((-1, 1))
# Fazendo a predição de km rodados com base nos novos valores
# Isso é equivalente a calcular: model.intercept_ + model.coef_ * x
model.predict(new)

# Visualizando os dados e a linha de regressão
plt.scatter(x, y)  # Pontos de dados originais
plt.plot(x, model.predict(x), color='blue')  # Linha de regressão
plt.show()  # Exibe o gráfico
```

# Exemplo 2: Regressão linear múltipla

```python
# Importando pandas para manipulação de dados tabulares
import pandas as pd
from sklearn.linear_model import LinearRegression

# Carregando o dataset de um URL
df = pd.read_csv("https://www.w3schools.com/python/data.csv")

# Definindo as variáveis independentes (features): Peso e Volume do carro
X = df[['Weight', 'Volume']]
# Definindo a variável dependente (target): Emissão de CO2
y = df['CO2']

# Criando o modelo de regressão
regr = LinearRegression()
# Treinando o modelo com os dados
regr.fit(X, y)

# Criando um DataFrame para predição com as mesmas colunas dos dados de treino (Weight e Volume)
# Aqui prevemos para um carro com 2300kg e volume de 1300cm³
new_predict = pd.DataFrame({'Weight': [2300], 'Volume': [1300]})
# Fazendo a predição da emissão de CO2 para o novo carro
predictedCO2 = regr.predict(new_predict)

# Exibindo o resultado da predição
print(predictedCO2)
```
